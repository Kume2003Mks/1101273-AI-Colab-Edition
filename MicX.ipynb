```python
from sklearn.feature_extraction.text import CountVectorizer #แปลงคำเป็น 0 or 1
from pythainlp import sent_tokenize, word_tokenize #แยกคำไทย

#connect google drive
from google.colab import drive
drive.mount('/content/drive')
!ls "/content/drive/My Drive/Colab/AI73/"
```

```python
import pandas as pd #DataFrame
dfSongName = pd.read_excel('/content/drive/My Drive/Colab/AI73/output1.xlsx') #อ่านไฟล์ เพลง + label
dfSongName.fillna('', inplace=True)
dfSongName
```

```python
#input เนื้อเพลง
if str(dfSongName.iloc[-1,1]) =='nan':
  dfSongName = dfSongName.drop([len(dfSongName)-1])

inpWord = str(input("ใส่เนื้อเพลงที่นี่: "))
print(inpWord)

#สร้าง dataframe เก็บเนื้อเพลง
inp_Lyrics = pd.DataFrame([inpWord],columns =['Lyrics'])

#รวมข้อมูล เก่า + ใหม่
dfSongName = pd.concat([dfSongName,inp_Lyrics],ignore_index = True)
dfSongName
```

```python
#function แยกคำ 
def my_tokenise(txt):
  txt = word_tokenize(txt, keep_whitespace=False)
  return ' '.join(txt)
```

```python
ForWord = pd.DataFrame() #เก็บ token
ForWord['token'] = dfSongName['Lyrics'].apply(my_tokenise) #แยกคำทุก column
ForWord
```

```python
ForWord.iloc[-1,-1] # check last column input
```

```python
#แปลงข้อมูลในรูป 0,1
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(ForWord.loc[:,'token'])

dfX = pd.DataFrame(X.toarray()) #เก็บไฟล์ที่แปลงแล้ว
dfX['Label'] = dfSongName.loc[:,'Label'] #เพิ่ม column 'Label'
dfX 
```

#Random forest
**Split training & test sets**

```python
X = dfX.iloc[:-1,dfX.columns!='Label'] #ทุก columns ยกเว้น row ล่าสุด
y = dfX.iloc[:-1,dfX.columns == 'Label'] #columns 'Label' ยกเว้น row ล่าสุด
  
#split&training sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

from sklearn.ensemble import RandomForestClassifier

# สร้าง random forset classifier
clf = RandomForestClassifier(n_estimators = 10)  
  
# train model
clf.fit(X_train, y_train)
  
# predictions on the test dataset
y_pred = clf.predict(X_test)
  
# metrics are used to find accuracy 
from sklearn import metrics  
from sklearn.metrics import classification_report

# accuracy calculation
print("ACCURACY OF THE MODEL: %.2f"% metrics.accuracy_score(y_test, y_pred))
```

```python
def chk_predict(inp_chk):
  if (inp_chk == 'T') or (inp_chk == 't'):
    return 'ถูกต้อง(correct)'
  elif (inp_chk == 'F') or (inp_chk == 'f'):
    return 'ผิด(wrong)'
  else:
    return 'กรุณากรอกข้อมูลให้ถูกต้อง'

def possible_Sum(predictions):
  arr = [['รัก',predictions[0,0]*100], ['สนุกสนาน',predictions[0,1]*100],['เศร้า',predictions[0,2]*100], ['ให้กำลังใจ',predictions[0,3]*100]]
  return arr
```

#ทำนาย และ บันทึก


```python
predictions = clf.predict_proba(dfX.iloc[-1:,dfX.columns != 'Label']) #prediction all
predictLabel = clf.predict(dfX.iloc[-1:,dfX.columns != 'Label']) #prediction value
print("ผลการทำนาย คือ ",predictLabel)
print("ความเป็นไปได้ทั้งหมด คือ ",possible_Sum(predictions))
chkloop = 0
while 1 >= chkloop:

  inp_store = input("\nบันทึกข้อมูลเพื่อใช้ในการ train model หรือไม่? (T/F) = ")

  if inp_store == 't' or inp_store == 'T':
      dfSongName.iloc[-1:,-1] = predictLabel
      print("\nบันทึกข้อมูลเรียบร้อย")
      dfSongName.to_excel('output1.xlsx')
      chkloop = 2
  elif inp_store == 'f' or inp_store == 'F':
      dfSongName = dfSongName.drop([len(dfSongName)-1])
      print("\nคุณเลือกไม่บันทึกข้อมูล T_T")
      chkloop = 2
  else :
      chkloop = 0
      print("กรุณากรอกข้อมูลให้ถูกต้อง")
```

```python
# from unidecode import unidecode
import re
from PIL import Image
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
```

```python
matplotlib.font_manager
matplotlib.rc("font", family="Itim-Regular")

font_path = '/content/drive/My Drive/Colab/AI73/Itim-Regular.ttf'
```

```python
#word cloud เพลงเศร้า
stopwords = set(STOPWORDS)
stopwords.add("yeah")
stopwords.add("Da")
stopwords.add("Ooowee")
stopwords.add("Oh")
stopwords.add("Baby")
stopwords.add("u")
stopwords.add("One")

wordcloud = WordCloud(
    font_path=font_path,
    background_color = 'black',
    regexp=r"[\u0E00-\u0E7Fa-zA-Z']+",
    width = 1000,
    height = 500,
    max_words=200,
    stopwords = stopwords,
    collocations=False,).generate(' '.join(dfSongName.loc[(dfSongName.loc[:,"Label"] == "เศร้า"),"Lyrics"]))

fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
