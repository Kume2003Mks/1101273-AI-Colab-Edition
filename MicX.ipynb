```python
import pandas as pd #DataFrame
import re #Regex 
from sklearn.feature_extraction.text import CountVectorizer #แปลงคำเป็น 0 or 1
from pythainlp import sent_tokenize, word_tokenize #แยกคำไทย
```

```python
dfSongName = pd.read_excel('/content/drive/My Drive/AI70/ai73/Lyric_LabelNewClean.xlsx') #อ่านไฟล์ เพลง + label
```

```python
dfSongName
```

```python
#function แยกคำ 
def my_tokenise(txt):
  txt = word_tokenize(txt, keep_whitespace=False)
  return ' '.join(txt)
```

```python
#input เนื้อเพลง
inpWord = str(input("ใส่เนื้อเพลงที่นี่: "))
print(inpWord)

#สร้าง dataframe เก็บเนื้อเพลง
inp_Lyrics = pd.DataFrame([inpWord],columns =['Lyrics'])

#รวมข้อมูล เก่า + ใหม่
dfSongName = pd.concat([dfSongName,inp_Lyrics],ignore_index = True)
dfSongName
```

```python
ForWord = pd.DataFrame() #เก็บ token
ForWord['token'] = dfSongName['Lyrics'].apply(my_tokenise) #แยกคำทุก column
ForWord
```

```python
 ForWord.iloc[-1,-1] # test column input
```

```python
#แปลงข้อมูลในรูป 0,1
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(ForWord.loc[:,'token'])

print(X.shape)
```

```python
dfX = pd.DataFrame(X.toarray()) #เก็บไฟล์ที่แปลงแล้ว
dfX['Label'] = dfSongName.loc[:,'Label'] #เพิ่ม column 'Label'
dfX 
```

#Random forest
**Split training & test sets**

```python
X = dfX.iloc[:-1,dfX.columns!='Label'] #ทุก columns ยกเว้น columns ล่าสุด
y = dfX.iloc[:-1,dfX.columns == 'Label'] #columns 'Label' ยกเว้น columns ล่าสุด
  
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

from sklearn.ensemble import RandomForestClassifier

# สร้าง random forset classifier
clf = RandomForestClassifier(n_estimators = 10)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
  
# performing predictions on the test dataset
y_pred = clf.predict(X_test)
  
# metrics are used to find accuracy or error
from sklearn import metrics  

# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
def chk_predict(inp_chk):
  if (inp_chk == 'T') or (inp_chk == 't'):
    return 'ถูกต้อง(correct)'
  elif (inp_chk == 'F') or (inp_chk == 'f'):
    return 'ผิด(wrong)'
  else:
    return 'กรุณากรอกข้อมูลให้ถูกต้อง'

def possible_Sum(predictions):
  arr = [['รัก',predictions[0,0]*100], ['สนุกสนาน',predictions[0,1]*100],['เศร้า',predictions[0,2]*100], ['ให้กำลังใจ',predictions[0,3]*100]]
  return arr
```

```python
predictions = clf.predict_proba(dfX.iloc[-1:,dfX.columns != 'Label'])
predictLabel = clf.predict(dfX.iloc[-1:,dfX.columns != 'Label'])
print("ผลการทำนาย คือ ",predictLabel)
print("ความเป็นไปได้ทั้งหมด คือ ",possible_Sum(predictions))

inp_chk = input("\nข้อมูลที่ได้คุณเห็นด้วยหรือไม่(T/F) = ")
print("\nresult is = %s"%chk_predict(inp_chk))

if (inp_chk == 't') or (inp_chk == 'T'):
  print("\nเย้ๆๆๆ กระผมดีใจที่ได้ช่วยเหลือครับ😀😉😄")
elif (inp_chk == 'f') or (inp_chk == 'F'):
  print("\nฮือๆๆ กระผมยังไม่เก่งเลยขอโทษนะครับ😭😭")
else:
  print("\n กรอกข้อมูลให้ถูกด้วยคร้บป๋ม😤")
```
